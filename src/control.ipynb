# 制御とは

時間とともに何かの法則に従って変化する系に対して、外部から操作を加えることによって、その系の状態の一部をある目的に適合するようにすることを、 **制御する** といいます。制御システムの話をしているときは、ここでの外部からの操作を **入力(input)** と呼び、操作したい状態を **出力(output)** と呼びます。

たとえば、以下は制御の例です。

* ロボットの姿勢を操作して倒れないようにする
* クレーンを操作して、所定の位置でぴったり止める
* タンクに水をためて、水位をあるラインに持っていく
* エアコンが、部屋の温度を快適に保つ
* 車を走らせて、白線の前でぴったり止める

## 常微分方程式による制御対象のシステムのモデル化

ここでは制御対象の状態が入力を含む、常微分方程式で表されるという仮定をおきます。
これは制御工学（とくに現代制御）では一般的な仮定です。

いくつか具体例を挙げます。

### マス-ばねシステム

質量 $M$ [kg] の物体が、ばね定数 $K$ [N/m] のばねにつながれていてるものとします。時刻 $t$ において、物体のつりあいの位置からの変位を $y(t)$ [m] とし、物体に外部から加える力を $f(t)$ [N] とし、物体と床面などとの摩擦は無視できるほど小さいとします。

このとき、物体とばねを含むシステム全体の状態は、次の運動方程式に従います。

$$
M \frac{d²y}{dt²} (t) = -K y (t) + f(t)
$$

### タンクシステム

円柱型のタンク [^tank] を考えます。タンクにはタンクに水を入れるパイプと、タンクから水を排出するパイプがあるものとして、タンクの断面積を $C$ [m²] とします。
そして時刻 $t$ において、流入する水量が $q_i(t)$ [m³/s] で、排出される水量が $q_o(t)$ [m³/s] で、タンクの水位が $h(t)$ [m] で表されるものとします。

このとき、システムの全体は以下の微分方程式に従います。

$$
C \frac{dh}{dt} (t) = q_i(t) - q_o(t)
$$

## 状態空間表現

システムで注目される量 [^state] を **状態変数(state variable)** と呼びます。
状態変数という言葉を使うと、いま例に挙げたシステムは「状態変数についての微分方程式で表せる」ことになります。
状態変数が複数あるときは、まとめて一つのベクトルだと思えばよいです。

一般に、システムを状態変数に関する微分方程式で表したものを **状態空間表現(state space representation)** と呼びます。
状態空間表現は、状態が時刻と制御入力に応じてどのように変化するかを表す **状態方程式(state equation)** と、出力が時刻と状態からどのように決まるかを表す **出力方程式(output equation)** からなります。

状態変数のベクトルを $𝐱(t)$ とし、制御入力のベクトルを $𝐮(t)$ とし、出力のベクトルを $𝐲(t)$ とすると、状態空間表現は一般にある関数 $f, g$ を用いて以下のように書けます。

$$
\begin{aligned}
\frac{d𝐱}{dt} (t) &= f(𝐱(t), 𝐮(t), t) \\
𝐲(t) &= g(𝐱(t), t)
\end{aligned}
$$

### マス-ばねシステム

二階の微分方程式ですが、状態変数として $𝐱(t) := (y(t), y'(t))$ とすれば一階の微分方程式に帰着することができて、状態空間表現の形になります。
出力ベクトル $𝐲(t)$ は変位 $y(t)$ そのものです。

状態方程式は次のように得られます。

$$
f(𝐱(t), 𝐮(t), t) = \left( y'(t), \frac{ - K y(t) + f(t) }{ M } \right)
$$

[^tank]: 液体や気体を入れる容器のこと。戦車のことではありません。
[^state]: 時刻に依存しない量は固定されていて制御の対象にならないので、システムで注目される量は時刻に依存しています。
